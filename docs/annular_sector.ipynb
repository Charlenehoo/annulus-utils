# 计算由两个同心圆和两条竖直线围成的第一象限区域面积

## 问题描述

设平面区域 $S(r, R, b)$ 由以下四条曲线围成：
- 两个同心圆：$x^2 + y^2 = r^2$ 和 $x^2 + y^2 = R^2$，其中 $0 < r < R$；
- 两条竖直线：$x = 0$（即 $y$ 轴）和 $x = b$，其中 $0 < b < r$；
- 并且限定在第一象限内，即 $x > 0$，$y > 0$。

求该区域的面积 $S(r, R, b)$。

## 几何分析

由于 $x$ 的取值范围是从 $0$ 到 $b$，且 $b < r$，因此对于每个固定的 $x \in [0, b]$，$y$ 的取值范围是从小圆的上半圆弧 $y = \sqrt{r^2 - x^2}$ 到大圆的上半圆弧 $y = \sqrt{R^2 - x^2}$。区域左侧边界为 $x = 0$ 上的线段从 $(0, r)$ 到 $(0, R)$，右侧边界为 $x = b$ 上的线段从 $(b, \sqrt{r^2 - b^2})$ 到 $(b, \sqrt{R^2 - b^2})$。这是一个曲边四边形。

## 辅助函数定义

首先考虑一个更简单的区域：由单个圆 $x^2 + y^2 = a^2$、直线 $y = 0$、$x = 0$、$x = b$ 在第一象限内围成的曲边梯形（其中 $0 < b < a$）。记其面积为 $A(a, b)$，则
$$
A(a, b) = \int_0^b \sqrt{a^2 - x^2} \, dx.
$$

## 面积计算

观察原区域，它恰好是大圆下的曲边梯形 $A(R, b)$ 减去小圆下的曲边梯形 $A(r, b)$，即
$$
S(r, R, b) = A(R, b) - A(r, b).
$$

下面分别计算 $A(R, b)$ 和 $A(r, b)$。利用积分公式
$$
\int \sqrt{a^2 - x^2} \, dx = \frac{x}{2} \sqrt{a^2 - x^2} + \frac{a^2}{2} \arcsin \frac{x}{a} + C,
$$
得
$$
A(R, b) = \int_0^b \sqrt{R^2 - x^2} \, dx = \frac{b}{2} \sqrt{R^2 - b^2} + \frac{R^2}{2} \arcsin \frac{b}{R},
$$
$$
A(r, b) = \int_0^b \sqrt{r^2 - x^2} \, dx = \frac{b}{2} \sqrt{r^2 - b^2} + \frac{r^2}{2} \arcsin \frac{b}{r}.
$$

因此
$$
S(r, R, b) = \left( \frac{b}{2} \sqrt{R^2 - b^2} + \frac{R^2}{2} \arcsin \frac{b}{R} \right) - \left( \frac{b}{2} \sqrt{r^2 - b^2} + \frac{r^2}{2} \arcsin \frac{b}{r} \right).
$$

整理得
$$
S(r, R, b) = \frac{b}{2} \left( \sqrt{R^2 - b^2} - \sqrt{r^2 - b^2} \right) + \frac{R^2}{2} \arcsin \frac{b}{R} - \frac{r^2}{2} \arcsin \frac{b}{r}.
$$

## 最终结果

$$
\boxed{S(r, R, b) = \dfrac{b}{2}\left( \sqrt{R^2 - b^2} - \sqrt{r^2 - b^2} \right) + \dfrac{R^2}{2}\arcsin\dfrac{b}{R} - \dfrac{r^2}{2}\arcsin\dfrac{b}{r}}
$$

其中参数满足 $0 < b < r < R$。

```python
import math


def area_under_circle(radius: float, b: float) -> float:
    """
    计算由圆 x^2 + y^2 = radius^2、直线 y = 0、x = 0、x = b
    在第一象限内围成的曲边梯形面积。

    参数:
        radius: 圆的半径，必须为正且大于 b
        b:      x 坐标右边界，满足 0 < b < radius

    返回:
        曲边梯形面积，公式为 (b/2)*sqrt(radius^2 - b^2) + (radius^2/2)*arcsin(b/radius)
    """
    if not (0 < b < radius):
        raise ValueError("参数必须满足 0 < b < radius")

    term1 = (b / 2) * ((radius**2 - b**2) ** 0.5)
    term2 = (radius**2 / 2) * math.asin(b / radius)
    return term1 + term2


def area_annular_sector(r: float, R: float, b: float) -> float:
    """
    计算由两个同心圆 (x^2+y^2=r^2 和 x^2+y^2=R^2) 与两条竖直线 (x=0, x=b)
    在第一象限内围成的扇环形面积。

    参数会自动识别内外半径：内圆半径为 min(r,R)，外圆半径为 max(r,R)，
    要求 0 < b < min(r,R)。

    参数:
        r, R: 两个圆的半径（顺序无关，自动交换）
        b:    x 坐标右边界

    返回:
        扇环形区域的面积
    """
    inner = min(r, R)
    outer = max(r, R)

    if not (0 < b < inner):
        raise ValueError("参数必须满足 0 < b < 内圆半径，且内圆半径小于外圆半径")

    return area_under_circle(outer, b) - area_under_circle(inner, b)
```

# 计算直线 $x = b$ 被两个同心圆所截的线段长度（第一象限）

## 问题描述

在第一象限内，给定两个同心圆：
- 内圆：$x^2 + y^2 = r^2$，半径 $r > 0$；
- 外圆：$x^2 + y^2 = R^2$，半径 $R > r$。

以及一条竖直线 $x = b$，其中 $0 < b < r$（保证直线与两个圆都相交）。求该直线被两个圆所截得的位于第一象限内的线段长度。

## 几何分析

由于直线 $x = b$ 与两个圆相交，交点坐标满足：
- 与外圆交点：$y = \sqrt{R^2 - b^2}$（取正值，因第一象限 $y>0$）；
- 与内圆交点：$y = \sqrt{r^2 - b^2}$（取正值）。

这两个交点之间的竖直线段即为所求，其长度等于两交点纵坐标之差。

## 长度计算

线段长度 $L$ 为：
$$
L = \sqrt{R^2 - b^2} - \sqrt{r^2 - b^2}.
$$

由于 $R > r$ 且 $b < r$，被开方数均为正，长度为正。

## 最终结果

$$
\boxed{L = \sqrt{R^2 - b^2} - \sqrt{r^2 - b^2}}
$$

其中参数满足 $0 < b < r < R$。

```python
def line_segment_length(r: float, R: float, b: float) -> float:
    """
    计算直线 x = b 在第一象限内被两个同心圆 x^2+y^2=r^2 和 x^2+y^2=R^2 所截的线段长度。

    参数:
        r, R: 两个圆的半径（顺序无关，函数会自动将较小的作为内圆，较大的作为外圆）
        b:    直线 x = b 的 x 坐标，必须满足 0 < b < min(r, R)

    返回:
        线段长度 L = sqrt(max(r,R)^2 - b^2) - sqrt(min(r,R)^2 - b^2)

    异常:
        ValueError: 如果参数不满足 0 < b < min(r, R)
    """
    inner = min(r, R)
    outer = max(r, R)

    if not (0 < b < inner):
        raise ValueError("参数必须满足 0 < b < 内圆半径")

    return (outer**2 - b**2) ** 0.5 - (inner**2 - b**2) ** 0.5
```
